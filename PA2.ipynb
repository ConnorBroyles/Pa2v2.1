```python
age = int(input("What is the subject's age? "))

for x in range(age):
    age = int(input("What is the subject's age? "))
    if age<18 and age>0:
        print("Minor")
        continue
    if age<64 and age>18:
        print("Adult")
        continue
    if age>=64:
        print("Senior Citizen")
        continue
    if age==0:
        break
```
